```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ryupepa/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df
```

```python
model = smf.ols(data=df, formula='output~capital+labour+wage').fit()
hat_y=model.predict()
```

```python
hat_y
```

```python
df['out2'] = hat_y**2
df['out3'] = hat_y**3
df['out4'] = hat_y**4
df
```

```python
model2=smf.ols(data=df, formula='output~capital+wage+labour+out2+out3+out4').fit()
model2.params
```

```python
R2_un = model2.rsquared
R2_re = model.rsquared
```

```python
F_obs=(R2_un-R2_re)/(1-R2_un)*(len(df)-7)/3
F_obs
```

```python
alpha = 0.05
stats.f.ppf(1-alpha,3,len(df)-7)
```
